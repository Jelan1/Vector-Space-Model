If I draw twice in a row from a bag that initially contained 2 green  marblea and 2 blue marbles, it's pretty hard to predict how I draw next. There are many different possibilities. 

If I draw with replacement, though, we can predict the odds off getting two greens is 
$$P({green∩}{green}) = P({green})*(P({green}) $$
$$P({green}_{1/2}*{green}_{1/2}) = 1/4 $$
and in general 
$$P({green∩}{blue}) = P({green})*P({blue}) $$

As long as the causality of the events are not intertwined, the joint probabilities 
$$P({green},{green}) $$
$$P({blue},{green}) $$
$$P({blue},{blue}) $$
remain a product of their marginal events
$$P({green}) = 1/2$$
$$P({blue}) = 1/2 $$
**We can say the chances of drawing a green *given* I just drew a green is the same as drawing two greens in a row.**
$$P({green}|{green}) = P({green}∩{green})$$
